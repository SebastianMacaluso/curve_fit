```python
fig2, (axes) = plt.subplots(nrows=1, ncols=1)
fig2.set_size_inches(6,5)
plt.tight_layout(pad=0.4, w_pad=5, h_pad=1.0)
markersize = 500

x=np.asarray([3000,4500,1800,4500])
y=np.asarray([2,6,3,3])
z=np.asarray([1500,1500,1200,2000])

for i in range(1):
    
     
#     x = np.linspace(min(results["gt_llh"]), max(results["trellis_MLE"]), 1000)
    cl1 = axes.scatter(x,y, c=z ,  marker="X", s=markersize, label="")
    
    cb = fig2.colorbar(cl1, ax=axes, fraction=0.15, shrink=1.0, aspect=20)
    cb.set_label(label="Costo?", size=25)
    cb.ax.tick_params(labelsize=15)
    

    axes.tick_params(labelsize=25) 
#     axes.plot(x, x, color="black", linestyle='--')
    axes.grid(which='both', axis='both', linestyle='--')
    axes.legend(loc='lower left', fontsize=25)
    
    axes.set_ylabel(r"# Escuelas", fontsize=30)
    axes.set_xlabel(r" # Alumnos", fontsize=30)
```

```python
import numpy as np
from scipy.optimize import curve_fit

def func(X, a, b):
    x,y = X
    return x**(a)/x * y**(b)/y

# some artificially noisy data to fit
x=np.asarray([3000,4500,1800,4500])
y=np.asarray([2,6,3,3])
z=np.asarray([1500,1500,1200,2000])
# a, b, c = 1/2, 3/4
# z = func((x,y), a, b, c) * 1 + np.random.random(101) / 100

# initial guesses for a,b,c:
p0 = 1/2, 3/4
ab, pcov = curve_fit(func, (x,y), z, p0)
```

```python
a,b = ab[0],ab[1]
```

```python
func((3000,2),a,b)
```

```python

from mpl_toolkits.mplot3d import Axes3D  

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
y = np.linspace(100,10000,200)
x = np.linspace(1,10,10)
X, Y = np.meshgrid(x, y)
zs = np.array(func((np.ravel(X), np.ravel(Y)),a,b))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('# Escuelas')
ax.set_ylabel('# Alumnos')
ax.set_zlabel('Costo?')

plt.show()
```
